```python
import re

# Define a function to extract the year from the 'Born' column
def extract_year(date):
    # Use regular expression to find patterns representing years
    match = re.search(r'\b\d{4}\b', str(date))
    # If a match is found, return the matched year
    if match:
        return match.group()
    else:
        return None

# Apply the function to the 'Born' column to extract years
data['Born'] = data['Born'].apply(extract_year)
```

```python
def extract_far(date):
  match = re.search(r'\b\d{4}\b', str(date))
  if match:
    return match.group()

  else:
    return None

data['Died']=data['Died'].apply(extract_far)
```
